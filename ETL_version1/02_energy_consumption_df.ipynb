##### ***Tabla source_country_electricity*** , parte consumption y production


**Pivoteo de columnas**

```python
# Normalizacion en nombres de columnas
energy_consumption_by_country = energy_consumption_by_country.rename(columns={'other_renewable_consumption':'other_renewables_consumption','other_renewable_electricity':'other_renewables_electricity'})
```

```python
energy_consumption_by_country.head()
```

```python
# Eliminacion de columnas no involucradas
population_source_electricity = energy_consumption_by_country.drop(['population','carbon_intensity_elec','electricity_demand','electricity_generation','energy_per_capita','greenhouse_gas_emissions'], axis=1)
```

```python
population_source_electricity.head()
```

```python
# Cantidad de fuentes de produccion electrica
n_sources = len(energy_consumption_by_country.columns[energy_consumption_by_country.columns.str.match(r'(.*_elec_per_capita$)')])
n_sources
```

Tipos de fuentes de energia encontradas

```python
sources = ['coal', 'gas', 'hydro', 'nuclear', 'oil', 'other_renewables', 'renewables', 'solar', 'wind']
```

Columnas de sufijos repetidos: '_consumption', '_elec_per_capita', '_electricity', '_energy_per_capita'

```python
cols = ['country','year', 'source', 'consumptionElectricity', 'productionElectricity', 'energyPerCapita', 'elecPerCapita']
```

Uniendo todo en un DataFrame central, sin repetición de columnas

```python
base_df = pd.DataFrame(columns = ['country','year', 'source', 'consumptionElectricity', 'productionElectricity', 'energyPerCapita', 'elecPerCapita'])
```

```python
for n in sources:
    agg = population_source_electricity[['country', 'year', f"{n}_consumption", f"{n}_electricity", f"{n}_energy_per_capita", f"{n}_elec_per_capita"]]
    agg.columns = ['country','year','consumptionElectricity', 'productionElectricity', 'energyPerCapita', 'elecPerCapita']
    agg.insert(2, 'source', n)
    base_df = base_df.append(agg)
```

```python
base_df.head()
```

```python
base_df.drop_duplicates(subset=['country', 'year', 'source'], inplace=True)
base_df.sort_values(by=['country','year','source'], inplace=True)
```

```python
base_df
```

```python
base_df.reset_index(drop=True).to_csv('../normalizedDataset_2/source_country_electricity_consumption.csv', header=True, index=False)
```
